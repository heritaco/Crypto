```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data
data = pd.read_csv("data_crypto.csv", index_col="timestamp", parse_dates=True)
```

```python
# Calculate the correlation matrix
corr = data.corr()
```

```python
# Create a heatmap
plt.figure(figsize=(30, 30))
sns.heatmap(corr, annot=True, cmap='viridis', fmt='.2f', annot_kws={"size": 8, "family": "Verdana"})
plt.title('\nCorrelation Map of Crypto Data\n', fontsize=40, fontweight='bold', family='Verdana')

# Save the plot as a PDF file
plt.savefig("heatmap.pdf", format='pdf', bbox_inches='tight')

plt.show()
```
