```python
import pandas as pd
```

```python
import plotly.graph_objects as go
from plotly.offline import plot
```

```python
# Load the data
data = pd.read_csv("C:\\Heri\\GitHub\\Crypto\\02-Coingecko\\data-crypto.csv", index_col="timestamp", parse_dates=True)
```

```python
# Calculate the correlation matrix
corr = data.corr()
```

```python
fig = go.Figure(
    data=go.Heatmap(
        z=corr.values,
        x=corr.columns,
        y=corr.index,
        colorscale='Viridis',
        text=[[f"{val:.2f}" for val in row] for row in corr.values],
        texttemplate="%{text}",
        textfont={"size": 5, "color": "black"}
    )
)

fig.update_layout(
    title={
        'text': 'Correlation Map of Crypto Data',
        'font': {
            'size': 24
        }
    },
    autosize=False,
    width=1000,
    height=1000,
    margin=dict(l=100, r=100, t=100, b=100)
)

fig.update_xaxes(tickfont=dict(size=8))
fig.update_yaxes(tickfont=dict(size=8))
```

```python
plot(fig, filename='Heatmap.html')
```
